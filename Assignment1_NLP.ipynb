```python
import numpy as np

def numpy_boundary_transversal(matrix):
    """
    Traverse a 2D matrix in spiral order (clockwise from outside to inside)
    """
    if matrix.size == 0:
        return []

    rows, cols = matrix.shape
    result = []

    # Define boundaries
    top, bottom = 0, rows - 1
    left, right = 0, cols - 1

    while top <= bottom and left <= right:
        # Traverse right along top row
        for j in range(left, right + 1):
            result.append(matrix[top][j])
        top += 1

        # Traverse down along right column
        for i in range(top, bottom + 1):
            result.append(matrix[i][right])
        right -= 1

        # Traverse left along bottom row (if we still have rows)
        if top <= bottom:
            for j in range(right, left - 1, -1):
                result.append(matrix[bottom][j])
            bottom -= 1

        # Traverse up along left column (if we still have columns)
        if left <= right:
            for i in range(bottom, top - 1, -1):
                result.append(matrix[i][left])
            left += 1

    return result
```

```python
def numpy_alternate_sort(arr):
    arr_sorted = np.sort(arr)
    result = np.zeros_like(arr_sorted)

    left = 0
    right = len(arr_sorted) - 1

    for i in range(len(arr_sorted)):
        if i % 2 == 0:
            result[i] = arr_sorted[left]
            left += 1
        else:
            result[i] = arr_sorted[right]
            right -= 1

    return result
```

```python
import pandas as pd
data = {
    'Name': ['Alice', 'Bob', 'Charlie','John','Michael','Smith','Cooper','James','Richard','Clarke'],
    'Subject': ['Eng', 'Phy', 'Chem','Math','Sst','Eng','Phy','Chem','Math','Sst'],
    'Score': np.random.randint(low=50,high=100,size=10),
    'Grades': np.nan
}
df = pd.DataFrame(data)
```

```python
df
```

```python
def grade_allocation(row):
    if row['Score'] >= 90:
        return 'A'
    elif row['Score'] >= 80:
        return 'B'
    elif row['Score'] >= 70:
        return 'C'
    elif row['Score'] >= 60:  # Fixed: was df['Grades']
        return 'D'
    else:
        return 'F'

# Apply the function
df['Grades'] = df.apply(grade_allocation, axis=1)
```

```python
df
```

```python
df=df.sort_values('Score',ascending=False)
```
